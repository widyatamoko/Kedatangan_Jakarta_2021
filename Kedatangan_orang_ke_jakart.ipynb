```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df_jan=pd.read_csv('/content/drive/MyDrive/dataset/Kedatangan Jakarta/januari.csv')
df_feb=pd.read_csv('/content/drive/MyDrive/dataset/Kedatangan Jakarta/februari.csv')
df_mar=pd.read_csv('/content/drive/MyDrive/dataset/Kedatangan Jakarta/maret.csv')
df_apr=pd.read_csv('/content/drive/MyDrive/dataset/Kedatangan Jakarta/april.csv')
df_mei=pd.read_csv('/content/drive/MyDrive/dataset/Kedatangan Jakarta/mei.csv')
```

```python
kedatangan2021= pd.concat([df_jan,df_feb,df_mar,df_apr,df_mei])
```

```python
sns.distplot(kedatangan2021['jumlah'], fit=norm)
```

```python
kedatangan2021['jumlah'].describe()
```

```python
var = 'kecamatan'
data = pd.concat([kedatangan2021['jumlah'], kedatangan2021[var]], axis=1)
my_order = kedatangan2021.groupby(by=[var])["jumlah"].median().sort_values().iloc[::-1].index
f, ax = plt.subplots(figsize=(16,9))
fig=sns.boxplot(x=var,y='jumlah', data=data, order=my_order)
fig.axis()
plt.xticks(rotation=90)
```

```python
var = 'kelurahan'
data = pd.concat([kedatangan2021['jumlah'], kedatangan2021[var]], axis=1)
my_order = kedatangan2021.groupby(by=[var])["jumlah"].median().sort_values().iloc[::-1].index
f, ax = plt.subplots(figsize=(52,18))
fig=sns.boxplot(x=var,y='jumlah', data=data, order=my_order)
fig.axis()
plt.xticks(rotation=90)
```

```python
var = 'bulan'
data = pd.concat([kedatangan2021['jumlah'], kedatangan2021[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig=sns.boxplot(x=var,y='jumlah', data=data)
fig.axis()
plt.xticks(rotation=90)
```

```python
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})


# Initialize the FacetGrid object
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
kecamatan = sns.FacetGrid(kedatangan2021, row="kecamatan", hue="kecamatan", aspect=15, height=.5, palette=pal)

# Draw the densities in a few steps
kecamatan.map(sns.kdeplot, "jumlah",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
kecamatan.map(sns.kdeplot, "jumlah", clip_on=False, color="w", lw=2, bw_adjust=.5)

# passing color=None to refline() uses the hue mapping
kecamatan.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)


# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)


kecamatan.map(label, "jumlah")

# Set the subplots to overlap
kecamatan.figure.subplots_adjust(hspace=-.25)

# Remove axes details that don't play well with overlap
kecamatan.set_titles("")
kecamatan.set(yticks=[], ylabel="")
kecamatan.despine(bottom=True, left=True)
```
